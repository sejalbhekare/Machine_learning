# Droping unnessasary column 

```python
df.drop('mid', axis=1, inplace=True)
df.drop('runs_last_5', axis=1, inplace=True)
df.drop('wickets_last_5', axis=1, inplace=True)
df.drop('striker', axis=1, inplace=True)
df.drop('non-striker', axis=1, inplace=True)
```

# Separting Numeric and category Data

```python
df
```

```python
df_num=df.select_dtypes(['int','float'])
df_num
```

```python
df_object=df.select_dtypes(['object'])
df_object
```

# Using Label Encoder for converting category data into numeric

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df_object:
    df_object[i]=le.fit_transform(df_object[i]) 
#convert the cat to label encoder
df_object
```

# feacture scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
numerical = ["runs","wickets","overs"]
norm=StandardScaler()
df_num[numerical] = norm.fit_transform(df_num[numerical])
```

```python
df_num
```

# Concatenate Numerical and Categorical Columns

```python
df_new=pd.concat([df_num,df_object],axis=1)
df_new
```

# split the data

```python
y=df_new.iloc[:,3:4]
y
```

```python
x=df_new.drop('total',axis=1)
x
```

# Importing the models and Train them

```python
# Splitting the dataset into train and test sets: 70-30 split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=42)
```

```python
#importing required libraries 
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

#creating instance for the model
lr = LinearRegression()

# fit the model 
lr.fit(x_train, y_train)
```

```python
#computing the accuracy of the model performance
acc_train_lr = lr.score(x_train, y_train)
acc_train_lr*100
```

```python
acc_test_lr = lr.score(x_test, y_test)
acc_test_lr*100
```

```python
#predicting the target value from the model for the samples
y_test_lr = lr.predict(x_test)
y_train_lr = lr.predict(x_train)
```

```python
#computing root mean squared error (RMSE)
rmse_train_lr = np.sqrt(mean_squared_error(y_train, y_train_lr))
rmse_test_lr = np.sqrt(mean_squared_error(y_test, y_test_lr))

print('Linear Regression: The RMSE of the training set is:', rmse_train_lr)
print('Linear Regression: The RMSE of the testing set is:', rmse_test_lr)
```

# k-Nearest Neighbors Regression

```python
from sklearn.model_selection import GridSearchCV

# KNN Regression model
from sklearn.neighbors import KNeighborsRegressor

#creating instance for the model
knn = KNeighborsRegressor()
```

```python
# fit the model 
knn.fit(x_train, y_train)
```

```python
#computing the accuracy of the model performance
acc_train_knn = knn.score(x_train, y_train)
acc_train_knn
```

```python
acc_test_knn = knn.score(x_test, y_test)
acc_test_knn
```

```python
param_grid = {'n_neighbors':list(range(1, 31)), 'weights': ['uniform', 'distance'],'algorithm':['kd_tree', 'ball_tree', 'brute', 'auto']}
```

```python
#creating instance for the grid
knn_grid = GridSearchCV(knn, param_grid , cv=10)
```

```python
# fit the model 
knn_grid.fit(x_train, y_train)
```

```python
#Checking the best parameters for the model
knn_para = knn_grid.best_params_
print(knn_para)
```

```python
acc_train_knn=knn_grid.score(x_train,y_train)
acc_train_knn
```

```python
#computing the accuracy of the model performance
acc_test_knn = knn_grid.score(x_test, y_test)
acc_test_knn
```

```python
#predicting the target value from the model for the samples
y_test_knn = knn_grid.predict(x_test)
y_train_knn = knn_grid.predict(x_train)
```

```python
#computing root mean squared error (RMSE)
rmse_test_knn = np.sqrt(mean_squared_error(y_test, y_test_knn))

print('KNN Regression: The RMSE of the testing set is:', rmse_test_knn)
```
